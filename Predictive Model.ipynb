```python
#To extract only the churners from the data
churner = df.loc[df['Churn']==1]

#To extract only the non-churners from the data
nonchurner = df.loc[df['Churn']==0]
```

#### Identifying Features in Data and using Dummies to transform them in order to build the model

#### Create a list of 1 element containing the string 'purpose'. Call this list cat_feats

```python
cat_feats = ['MaritalStatus', 'Gender', 'Education', 'Occupation', 'CommuteDistance']
```

```python
final_data = pd.get_dummies(data2.drop(['CustomerKey','LastOrderDate'], axis=1),columns=cat_feats,drop_first=True)
```

#### Choosing a K Value
Let's go ahead and use the elbow method to pick a good K Value:

```python
error_rate = []

# Will take some time
for i in range(1,45):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_ktrain,y_ktrain)
    pred_i = knn.predict(X_ktest)
    error_rate.append(np.mean(pred_i != y_ktest))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,45),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

#### Here we can see that that after arouns K>32 the error rate just tends to hover around 0.36-0.37 Let's retrain the model with the lowest k=1 and check classification report

```python
knn = KNeighborsClassifier(n_neighbors=23)

knn.fit(X_ktrain,y_ktrain)
pred = knn.predict(X_ktest)

print('WITH K=1')
print('\n')
print(confusion_matrix(y_ktest,pred))
print('\n')
print(classification_report(y_ktest,pred))
```

Create an instance of the RandomForestClassifier class and fit it to our training data

#### Examine the Default Random Forest to Determine Parameters

```python
from pprint import pprint
# Look at parameters used by our current forest
print('Parameters currently in use:\n')
pprint(rfc.get_params())
```

#### Random Search with Cross Validation

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
pprint(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rfc = RandomForestClassifier()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rfc_random = RandomizedSearchCV(estimator = rfc, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rfc_random.fit(X_train,y_train)
```

#### Evaluation and prediction of the random search with cross validation model
